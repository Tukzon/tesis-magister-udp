```python
import pandas as pd
from ta import add_all_ta_features
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import yfinance as yf
import numpy as np
```

```python
ticker_symbol = "AAPL"  # Por ejemplo, AAPL para Apple Inc.

# Descargar los datos
df = yf.download(ticker_symbol, start="2014-01-01", end="2018-12-31")

df = df.reset_index()

df = add_all_ta_features(
    df, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True
)

# Mostrar los primeros registros
print(df.head())
```

```python
# Reemplazar valores infinitos con NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Reemplazar valores NaN con 0
df.fillna(0, inplace=True)

# Mostrar el DataFrame limpio
print("\nDataFrame limpio:")
print(df)
```

```python
import numpy as np

# Seleccionar las columnas que se quieren normalizar
columns_to_normalize = ['Open', 'High', 'Low', 'Close', 'Volume']

# Reemplazar valores nulos (NaN) e infinitos antes de la normalización
df[columns_to_normalize] = df[columns_to_normalize].replace([None, np.nan, float('inf'), float('-inf')], 0)

# Definir la función de normalización
def normalize(X, X_min, X_max):
    return (X - X_min) / (X_max - X_min)

# Aplicar la normalización a cada columna seleccionada
for column in columns_to_normalize:
    X_min = df[column].min()
    X_max = df[column].max()
    df[column] = normalize(df[column], X_min, X_max)

# Mostrar el DataFrame con los valores normalizados
print("\nDataFrame con valores normalizados:")
print(df)
```

```python
# Calculamos el índice de la fecha que marca el 80% del dataset
split_index = int(len(df) * 0.8)
split_date = df.index[split_index]

# Dividimos el dataset en train (80%) y test (20%)
train = df[df.index <= split_date]
test = df[df.index > split_date]

# Revisamos el tamaño de cada conjunto
print(f"Tamaño del conjunto de entrenamiento: {len(train)}")
print(f"Tamaño del conjunto de prueba: {len(test)}")
```

```python
# Adding Close_Shift Variable
train['Adj close anterior'] = train['Adj Close'].shift(1)
test['Adj close anterior'] = test['Adj Close'].shift(1)



# Adding Daily Returns variable
train['variacion'] = (train['Adj Close']/train['Adj close anterior'] - 1) * 100
test['variacion'] = (test['Adj Close']/test['Adj close anterior'] - 1) * 100

train['variacion'] = train['variacion'].fillna(0)
test['variacion'] = test['variacion'].fillna(0)
```

```python
train['clasificacion'] = np.where(train['variacion'] > 1, 1,
                 np.where(train['variacion'] < -1, -1, 0))

test['clasificacion'] = np.where(test['variacion'] > 1, 1,
                 np.where(test['variacion'] < -1, -1, 0))
```

```python
train = train.drop(columns=['Adj Close','Open','High','Low','Close','Volume','Adj close anterior','variacion'])
test = test.drop(columns=['Adj Close','Open','High','Low','Close','Volume', 'Adj close anterior','variacion'])
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score  # Importar f1_score

# Separar la variable objetivo (clasificación)
y_train = train['clasificacion']
y_test = test['clasificacion']

# Eliminar 'clasificacion' de las características (X)
X_train = train.drop(['clasificacion', 'Date'], axis=1)
X_test = test.drop(['clasificacion', 'Date'], axis=1)

# Reemplazar valores nulos (NaN), infinitos y negativos infinitos por 0
X_train = X_train.replace([None, np.nan, float('inf'), float('-inf')], 0)
X_test = X_test.replace([None, np.nan, float('inf'), float('-inf')], 0)

# Utilizar RFE con un modelo de Regresión Logística para la selección de características
model = LogisticRegression(penalty='l2', multi_class='multinomial', solver='lbfgs')
selector = RFE(model, n_features_to_select=5)
selector.fit(X_train, y_train)

# Obtener las 5 mejores características seleccionadas
selected_features = X_train.columns[selector.support_]
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Definir el tamaño de la ventana deslizante
window_size = 3

# Función para crear los datos con ventana deslizante usando las características seleccionadas
def create_sliding_window_data(data, window_size, selected_features):
    X, y = [], []
    for i in range(len(data) - window_size):
        window = data.iloc[i:i + window_size][selected_features]
        features = window.values  # Mantener la estructura 2D
        label = data.iloc[i + window_size]['clasificacion']
        X.append(features)
        y.append(label)
    return np.array(X), np.array(y)

# Crear los datos con la ventana deslizante para X_train y X_test
X_train_seq, y_train_seq = create_sliding_window_data(train, window_size, selected_features)
X_test_seq, y_test_seq = create_sliding_window_data(test, window_size, selected_features)

# Verificar el tamaño de la características de entrada
print(f"Tamaño de las características de entrada (X_train_seq): {X_train_seq.shape}")
print(f"Tamaño de las características de entrada (X_test_seq): {X_test_seq.shape}")

# Definir el modelo LSTM con una función para KerasClassifier
def create_lstm_model(neurons=50, optimizer='adam'):
    model = Sequential()
    model.add(LSTM(neurons, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Convertir el modelo a KerasClassifier para usarlo con GridSearchCV
model_lstm = KerasClassifier(model=create_lstm_model, verbose=0)

# Definir los parámetros para GridSearchCV
param_grid = {
    'model__neurons': [30, 50],  # Número de neuronas
    'batch_size': [16, 32],
    'epochs': [5, 10],
    'optimizer': ['adam', 'rmsprop']
}

# Lista para almacenar los resultados de accuracy, F1-score y los mejores hiperparámetros
accuracy_results = []
f1_results = []
best_params_list = []

# Ejecutar el proceso 20 veces
for i in range(20):
    # Definir GridSearchCV
    grid_search = GridSearchCV(estimator=model_lstm, param_grid=param_grid, n_jobs=-1, cv=3)

    # Entrenar el modelo con búsqueda de hiperparámetros
    grid_search.fit(X_train_seq, y_train_seq)

    # Obtener el mejor modelo y los mejores hiperparámetros
    best_params = grid_search.best_params_
    best_model = grid_search.best_estimator_

    # Hacer predicciones con el mejor modelo en el conjunto de prueba
    y_pred_prob = best_model.predict(X_test_seq)
    y_pred = (y_pred_prob > 0.5).astype(int)

    # Calcular la exactitud
    accuracy = accuracy_score(y_test_seq, y_pred)

    # Calcular el F1-score (puedes usar 'binary', 'macro', 'weighted', según el tipo de clasificación)
    f1 = f1_score(y_test_seq, y_pred, average='weighted')  # Cambiar 'weighted' si lo prefieres diferente

    # Almacenar la exactitud, el F1-score y los mejores hiperparámetros de la iteración actual
    accuracy_results.append(accuracy)
    f1_results.append(f1)
    best_params_list.append(best_params)

    # Imprimir el resultado de cada ejecución
    print(f"Exactitud en la ejecución {i + 1}: {accuracy:.4f}")
    print(f"F1-score en la ejecución {i + 1}: {f1:.4f}")
    print(f"Mejores hiperparámetros: {best_params}")

# Calcular y mostrar el promedio de las exactitudes y el F1-score
average_accuracy = np.mean(accuracy_results)
std_accuracy = np.std(accuracy_results)
average_f1 = np.mean(f1_results)
std_f1 = np.std(f1_results)

print(f"\nPromedio de la exactitud después de 20 ejecuciones: {average_accuracy:.4f}")
print(f"Desviación estándar de la exactitud después de 20 ejecuciones: {std_accuracy:.4f}")
print(f"Promedio del F1-score después de 20 ejecuciones: {average_f1:.4f}")
print(f"Desviación estándar del F1-score después de 20 ejecuciones: {std_f1:.4f}")

# Mostrar los mejores hiperparámetros en cada ejecución
print("\nMejores hiperparámetros en cada ejecución:")
for i, params in enumerate(best_params_list):
    print(f"Ejecución {i + 1}: {params}")
```
